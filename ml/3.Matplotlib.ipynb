# Matplotlib
- 저수준(세밀한) 그래프 그리기
- seaborn, pandas 그래프 기본

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline   # 그래프를 아래줄에 표시해주세용(안써도 괜춘 최근 버전 주피터)
```

```python
# graph 그리기
plt.figure()
plt.plot([1,2,3,4,5,6,7,8,9,8,7,6,5,4,3,2,1]) # 그리기
plt.show()   # 지금까지 그린 plot 한번에 보여줘
```

### **간단한 선 그래프 그리기**
- 보통은 2차원 그래프로 그린다.(3차원은 복잡해서 이해도 안됨)

```python
x = [1,2,3,4]
y = [2,3,4,5]
plt.plot(x,y)
plt.show()
```

### **배열을 이용하여 그래프 그리기**

```python
a = np.array([1,2,3,4])
b = np.array([2,3,4,5])
plt.plot(a,b)
plt.show()
```

### **Matplotlib로 기본차트 그리기**

```python
# numpy 를 이용한 sine 그래프
import numpy as np

t = np.arange(0, 12, 0.01)  # 0~12 까지 0.01간격으로 데이터 만들기
print(t)
y = np.sin(t)
print(y)
```

-  **캔버스 객체 생성** : 흰 도화지

```python
## 캔버스 객체 생성
# figsize : 사이즈
fig = plt.figure(figsize=(6,4))
```

- **axes 축 굵기**

```python
## axes 축에 대한 굵기를 진하게 처리
plt.rcParams['axes.linewidth'] = 2.0
```

```python
fig = plt.figure(figsize=(6,4))
plt.plot(t,y)  # t : 0~12 / y : sine
plt.show()
```

### **grid**

```python
plt.figure(figsize=(6,4))
plt.plot(t, y)
plt.grid()  # 그리드 적용
plt.show()
```

#### **x축, y축 라벨, 타이틀**

```python
plt.figure(figsize=(6,4))
plt.plot(t,y)
plt.grid()
plt.xlabel('time') # x축 라벨
plt.ylabel('Amplitude') # y축 라벨
plt.title('Example') # 타이틀
plt.show()
```

#### **2개 그래프 한 도화지에 같이 그리기**

```python
plt.figure(figsize=(6,4))
plt.plot(t, np.sin(t)) # sin 그래프
plt.plot(t, np.cos(t)) # cos 그래프추가
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example')
plt.show()
```

#### **legend(범례) 표시**
- 여러개 그래프 같이 그릴때 각 그래프가 뭔지 표시

```python
plt.figure(figsize=(6,4))
plt.plot(t, np.sin(t), label='sin')
plt.plot(t, np.cos(t), label = 'cos')
plt.grid()
plt.legend() # legend 표시
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example')
plt.show()
```

### **선의 굵기와 색상 지정**
- lw : 선 굵기
- color : 색상

```python
plt.figure(figsize=(6,4))
plt.plot(t, np.sin(t), lw=3, label='sin') # lw : 선의 굵기 조정
plt.plot(t, np.cos(t), color='r', label='cos') # color
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example')
plt.show()
```

### **x축과 y축 값 범위 지정**
- x limit
- y limit

```python
plt.figure(figsize=(6,4))
plt.plot(t, np.sin(t), lw=3, label = 'sin')
plt.plot(t, np.cos(t), 'r', label = 'cos')
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('example')
plt.ylim(-1.2, 1.2) # y 축 범위 지정
plt.xlim(0, np.pi) # x 축 범위 지정  
plt.show()
```

#### **그래프 좌표 점점 마다 레이블(마커) 지정**

```python
x = np.arange(1, 10)
y = x**2

fig = plt.figure(figsize=(8,4))
plt.plot(x, y , c='r', lw=3, ls='--', marker='s', ms=6, mec='b', mew=3, mfc='r')  # 마커 지정
plt.title('tick setting')
plt.xticks(x)
plt.yticks(y)
plt.show()
```

#### **서브 플롯 추가하기**

```python
# data
t = [0, 1, 2, 3, 4, 5, 6]
y = [1, 4, 5, 8, 9, 5, 3]

# graph
fig = plt.figure(figsize=(10, 6))

ax1 = fig.add_subplot(2, 2, 1) # 2행 2열 그래프의 첫번째 그래프
ax1.plot(t, y, color = 'green')


ax2 = fig.add_subplot(2,2,2) # 2행 2열 그래프의 2번째 그래프
ax2.plot(t, y, color='green', linestyle='dashed')

ax3 = fig.add_subplot(2,2,3)
ax3.plot(t, y, color = 'green', linestyle='dashed', marker='o')

ax4 = fig.add_subplot(2,2,4)
ax4.plot(t, y, color = 'green', linestyle='dashed', marker='o', markerfacecolor = 'blue')

plt.show()
```

#### **linestyle, marker 지정**

```python
plt.figure(figsize=(10, 6))
plt.plot(t, y , color='green', linestyle='dashed')  # linestyle='dashed' : line 
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, y ,color = 'green', linestyle='dashed', marker='o') # marker : 데이터 존재하는 곳에 마킹
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, y ,color='green', linestyle='dashed', marker='o', markerfacecolor = 'blue') # markerfacecolor, markersize : 마커 크기,색상 지정
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, y, color='green', linestyle = 'dashed', marker='o', markerfacecolor='blue', markersize=12)
plt.show()
```
