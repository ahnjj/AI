## Housing Price Prediction

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create two points(datas)
x_train = np.array([1.0, 2.0])  # 1000sqft
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

### m

```python
# m = the number of training examples
print(f'x_train.shape: {x_train.shape}')  # 2 
m = x_train.shape[0] # the length of the array
print(f"Number of training examples is : {m}")
```

```python
# find m using len()
m = len(x_train)
print(f"Number of training examples is : {m}")
```

### training example x_i, y_i

```python
i = 0

x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

### Plotting the data

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title('Housing Prices')
plt.ylabel('Price')
plt.xlabel('Size(1000sqft)')
plt.show()
```

### Model Function

```python
# Linear function
w = 100
b = 100
print(f'w: {w}')
print(f'b: {b}')
```

Compute the value of $f_{w,b}(x^{(i)})$ for two data points

$x^{(0)}$ : `f_wb = w * x[0] + b`

$x^{(1)}$ : `f_wb = w * x[1] + b`

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b) # predict y
 
plt.plot(x_train, tmp_f_wb, c='b', label='Prediction') # plot prediction 
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values') # plot the data points

plt.title('Housing Prices')
plt.ylabel('Price')
plt.xlabel('Size(1000sqft)')
plt.legend()
plt.show()
```
